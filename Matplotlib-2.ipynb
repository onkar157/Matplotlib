### Subplots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
#subplot
#plt.subplot(no.of rows, no.of.col, plot number)

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2)
```

```python
fig, ax = plt.subplots(nrows=3,ncols=3)
```

```python
fig, ax = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()                    #to fix the overlapping
```

```python
#It is array of matplotlib axes. It is just list of axes object. Means, we can actually iterate through it
ax
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)

for current_ax in ax:
    current_ax.plot(x,y)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
ax[0].plot(x,y)
ax[0].set_title('First_plot')
ax[1].plot(y,x)
ax[1].set_title('Second_plot')
plt.tight_layout()
plt.show()
```

### Figure size and DPI

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize=(8,2))
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4),dpi=100) #increase dpi for more detail
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
plt.show()
```

```python
#Save fig
fig.savefig('my_pic.png',dpi=200) #can change extension --> png,jpg etc
```

```python
#2 plots together and adding legend
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
#Legend

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x squared') #we need to add label for fun (ax.legend) to work.
ax.plot(x,x**3, label='x cubed')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x squared') #we need to add label for fun (ax.legend) to work.
ax.plot(x,x**3, label='x cubed')
ax.legend(loc=9)                #loc to avoid overlapping legend and graph.
                                #loc takes range from 0 to 10
```
