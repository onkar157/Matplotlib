```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Plot Appearance

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.show()
```

```python
# adding color

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple')
plt.show()
```

```python
#RGB hex code color

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00')
plt.show()
```

```python
# linewidth

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',linewidth=3) # lw will also work
plt.show()
```

```python
#Transparancy (alpha)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',linewidth=3,alpha=0.6)
plt.show()
```

```python
#Linestyle 

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',linewidth=3,alpha=1,linestyle='--')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',linewidth=3,alpha=1,linestyle='-.')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',lw=3,alpha=1,ls=':') #ls will also work
plt.show()
```

```python
#Markers (only use when we have few datatypes)

len(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',lw=3,alpha=1,ls=':', marker='o') 
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',lw=0.5,alpha=1,ls=':', marker='+') 
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00',lw=0.5,alpha=1,ls=':', marker='1',markersize=25) #changing marker size
plt.show()
```

```python
#marker-face-color

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple',lw=2,alpha=1,ls=':', marker='o',markersize=20,
       markerfacecolor='yellow') 
plt.show()
```

```python
#marker-edge-width

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple',lw=2,alpha=1,ls=':', marker='o',markersize=20,
       markerfacecolor='yellow',markeredgewidth=2) 
plt.show()
```

```python
#marker-edge-color

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple',lw=2,alpha=1,ls=':', marker='o',markersize=20,
       markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='blue') 
plt.show()
```

##### Control over axis appearance and plot range

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple',alpha=1,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
plt.show()
```
