```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
#Functional Method

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r-')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
#subplot
#plt.subplot(no.of rows, no.of.col, plot number)

plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
#object orianted method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
plt.show()
```

```python
fig = plt.figure()
#axes1 = fig.add_axes([left,bottom,width,height])
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes.set_title('Title')
plt.show()
```

```python
fig = plt.figure()
#axes1 = fig.add_axes([left,bottom,width,height])
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  #adding set of axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])  #all these values should between 0 to 1.

axes1.plot(x,y)
axes1.set_title('Large plot')
axes2.plot(y,x)
axes2.set_title('Small plot')
plt.show()
```
