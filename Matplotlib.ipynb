```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline       #plots staying in the notebook
plt.style.use("seaborn-darkgrid")  #to create background of the chart
```

```python
x = [-3,5,7]
y = [10,2,5]
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.scatter(x,y)
```

```python
plt.figure(figsize=(12,5))  #customising the size 
plt.plot(x,y)
plt.scatter(x,y)
```

```python
plt.figure(figsize=(12,5))  
plt.plot(x,y, color="green")
plt.scatter(x,y, color="red")
```

```python
plt.figure(figsize=(12,5))  
plt.plot(x,y, color="green", linestyle="--", linewidth="3") #changing line style n width
plt.scatter(x,y, color="red", s=80)   #s for changing size of the point
```

```python
plt.figure(figsize=(12,5),facecolor="lightblue")  #facecolor to add border to the map
plt.plot(x,y, color="green", linestyle="--", linewidth="3") 
plt.scatter(x,y, color="red", s=80) 
plt.title("Fist plot", color="blue", size=40) #setting color and font size for the title.
plt.xlabel("x-axis", color="blue", size=20)
plt.ylabel("y-axis", color="blue", size=20)
```

```python
plt.figure(figsize=(12,5),facecolor="lightblue")  
plt.plot(x,y, color="green", linestyle="--", linewidth="3") 
plt.scatter(x,y, color="red", s=80) 
plt.title("First plot", color="blue", size=40) 
plt.xlabel("x-axis", color="blue", size=20)
plt.ylabel("y-axis", color="blue", size=20)
plt.xticks(np.arange(-4,10,1))  #arranging x-axis 
plt.yticks(np.arange(1,12,1))
plt.savefig("demo.png", dpi=300)    #saving the plot
```

```python
plt.figure(figsize=(12,7), facecolor='lightgreen')
plt.scatter(df.hp,df.disp,s=20,color='red')
plt.xlabel("horsepower")
plt.ylabel("displacement")
plt.title("correlation plot")
```

```python
#Histplot
plt.figure(figsize=(12,7),facecolor="lightgreen")
plt.hist(df['hp'], facecolor='yellow', edgecolor='red', bins=5)
plt.xlabel("HP",size=30)
```

```python
#Boxplot
plt.figure(figsize=(12,5), facecolor='lightgreen')
plt.boxplot(df['hp'], vert=False)
```

```python
plt.figure(figsize=(12,5), facecolor='lightgreen')
plt.violinplot(df['hp'], vert=False)
plt.xlabel("HP")
```
